```python
def limpa_preco_ajusta_alugueis(linha):
    linha['price'] = linha['price'].replace('R$', '').replace('/mês', '').replace('.', '').replace(' ', '').strip()
    linha['price'] = int(''.join(filter(str.isdigit, linha['price'])))
    preco = int(linha['price'])
    if preco < 10000: 
        preco = preco * 200
    linha['price'] = preco
    return linha
```

```python
df.apply(lambda x: limpa_preco_ajusta_alugueis(x), axis = 1)

```
