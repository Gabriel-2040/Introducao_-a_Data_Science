```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'E:\PYTHON\CURSOS\11_SANTANDER_CODER_2024\Introducao_ a_Data_Science\biblioteca\netflix_titles.csv')
df = df.drop(df.columns[0],axis=1)
```

```python
a = df.head()
print(a)
```

```python
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']

movie_durations = movies['duration'].value_counts().sort_index()
tv_show_durations = tv_shows['duration'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(12, 6))

# Plot movie durations
plt.bar(movie_durations.index, movie_durations.values, label='Movies', alpha=0.7)

# Plot TV show durations
plt.bar(tv_show_durations.index, tv_show_durations.values, label='TV Shows', alpha=0.7)

# Add titles and labels
plt.title('Distribution of Durations for Movies and TV Shows')
plt.xlabel('Duration')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()

# Show the plot
plt.show()
# plt.bar(eixo_x, eixo_y)
# plt.title('Filmes Ano / tipo')
# plt.xlabel('aaa')
# plt.ylabel('bbbb')
# plt.xticks(rotation = 45)
# plt.show()
```

```python
dados_este_ano = {'mes': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
                  'valor_total': [1000, 1500, 1800, 1200, 1600, 1700, 1400, 1550, 1900, 2100, 2200, 2300]}
dados_ano_passado = {'mes': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
                     'valor_total': [900, 1400, 1700, 1100, 1500, 1600, 1300, 1450, 1800, 2000, 2100, 2200]}

df_este_ano = pd.DataFrame(dados_este_ano)
df_ano_passado = pd.DataFrame(dados_ano_passado)

# Plotando os dados
plt.figure(figsize=(10, 6))

plt.plot(df_este_ano['mes'], df_este_ano['valor_total'], label='Este Ano', marker='o')
plt.plot(df_ano_passado['mes'], df_ano_passado['valor_total'], label='Ano Passado', marker='o')

# Adicionando ttulos e legendas
plt.title('Comparao das Vendas Mensais: Este Ano vs. Ano Passado')
plt.xlabel('Ms')
plt.ylabel('Valor Total das Vendas')
plt.legend()

# Exibindo o grfico
plt.show()
```
